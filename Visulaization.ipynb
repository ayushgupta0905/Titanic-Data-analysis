```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("cleaned_data1.csv")
```

```python
df.info()
```

```python
count1= df["Sex"].value_counts()
plt.figure(figsize=(7,6))
plt.bar(count1.index,count1,color=["skyblue","Yellow"])
plt.title("People present on Titanic",fontweight="bold")
plt.xlabel("Gender")
plt.ylabel("Number of People")
plt.show()
```

**Insight:** The percentage of male is more than that of the female in the overall people.

```python
df["Pclass"]=df["Pclass"].sort_values()
count2 = df["Pclass"].value_counts()
labels=["Lower class (3)","Higher class (1)","Middle class (2)"]
plt.figure(figsize=(8,7))
plt.pie(count2,shadow ="True",labels=labels,startangle=90,wedgeprops={"edgecolor": "white", "linewidth": 0.5},autopct = "%1.1f%%")
plt.title("Passenger class",fontweight="bold")
plt.show()
```

**Insight:** The percentage of Lower class is more than together of middle and higher class.

```python
count3 = df["Survived"].value_counts()
plt.figure(figsize=(6,5))
plt.bar(["Died","Survived"], count3, color=["Red","Green"])
plt.title("Survival Count", fontweight="bold")
plt.xlabel("Outcome")
plt.ylabel("Number of People")
plt.show()
```

**Insight:** The plots gives that more than 500 people were died and nearly 300 could survive the incident.

```python
survival_sex = df.groupby("Sex")["Survived"].mean()
plt.figure(figsize=(6,5))
plt.bar(survival_sex.index, survival_sex.values, color=["skyblue","blue"])
plt.title("Survival Rate by Gender", fontweight="bold")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.ylim(0,1)
plt.show()
```

**Insight:**Women survived at a much higher rate than men which is about 74% of female passengers survived compared to only around 19% of male passengers. This is the strongest single predictor of survival in the dataset, likely reflecting the "women and children first" evacuation priority.

```python
survival_class = df.groupby("Pclass")["Survived"].mean()
plt.figure(figsize=(6,5))
plt.bar(survival_class.index.astype(str), survival_class.values, color="skyblue")
plt.title("Survival Rate by Passenger Class", fontweight="bold")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.ylim(0,1)
plt.show()
```

**Insight:**Survival rate drops clearly as class decreases, around 63% for 1st class, 47% for 2nd class, and only about 24% for 3rd class. This suggests wealthier, higher-class passengers had better access to lifeboats, likely due to cabin location and priority boarding.

```python
count4 = df["Embarked"].value_counts()
plt.figure(figsize=(6,5))
plt.bar(count4.index, count4.values, color="skyblue")
plt.title("Passengers by Port of Embarkation", fontweight="bold")
plt.xlabel("Port")
plt.ylabel("Number of People")
plt.show()
```

**Insight:**Most passengers boarded from Southampton (S), followed by Cherbourg (C), with far fewer boarding from Queenstown (Q). Southampton alone accounts for roughly three-quarters of all passengers.

```python
plt.figure(figsize=(8,6))
colors = df["Survived"].map({0:"Red", 1:"Green"})
plt.scatter(df["Age"], df["Fare"], c=colors, alpha=0.6, edgecolor="white", linewidth=0.3)
plt.title("Age vs Fare (colored by survival)", fontweight="bold")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

**Insight:**Survivors (green) are noticeably concentrated at higher fare values, while most low-fare passengers (bottom of the chart) did not survive (red).

```python
plt.figure(figsize=(7,5))
plt.hist(df["Age"], bins=30, color="skyblue", edgecolor="white")
plt.title("Age Distribution of Passengers", fontweight="bold")
plt.xlabel("Age")
plt.ylabel("Number of People")
plt.show()
```

**Insight:**Most passengers were young adults, with the distribution concentrated between roughly 20–35 years old.
